**Objetivo:**  
* Realizar un estudio estadístico descriptivo sobre los datos del archivo.
* Analizar las variables numéricas del conjunto de datos para comprender su distribución, tendencia central, dispersión y posibles valores atípicos.  

**Herramientas**:
* Lenguaje de programación: Python
* Librerías de Python: Pandas, Numpy, Seaborn y Matplotlib

**Pasos realizados:**  
1. **Estadísticos descriptivos básicos:**  
   - Se calcularon medidas como la **media, mediana, desviación estándar, mínimo, máximo y cuartiles** para cada columna numérica.  
   - Esto permite evaluar la tendencia central y la variabilidad de los datos.  

2. **Visualización de distribuciones:**  
   - **Histogramas:** Para observar la forma de la distribución (normal, sesgada, bimodal, etc.).  
   - **Diagramas de caja (box plots):** Para identificar valores atípicos y comparar la dispersión entre variables.  

3. **Análisis de valores faltantes:**  
   - Se verificó la presencia de datos faltantes o inconsistentes que puedan afectar el análisis.  

**Resultados esperados:**  
- Identificar qué variables siguen una distribución normal y cuáles presentan sesgos.  
- Detectar posibles errores o anomalías en los datos (valores extremos o faltantes).  
- Determinar si algunas columnas son categóricas (como códigos) en lugar de numéricas.  

**Ejemplo de interpretación:**  
Si una columna muestra una media y mediana similares con poca desviación estándar, su distribución sería cercana a la normal. En cambio, si tiene una mediana muy distinta de la media con una cola larga en el histograma, indicaría un sesgo, posiblemente por valores faltantes o concentración en extremos.  

Este estudio proporciona una base para futuros análisis, como comparaciones entre grupos o modelos predictivos.

## 3. Visualización de los datos

Generamos visualizaciones para explorar la distribución de los datos. Los gráficos seleccionados son histogramas y diagramas de caja (box plot).


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


num_cols = df.select_dtypes(include=['number'])
cols_excluir = ['Cod. Lingua Estranxeira', 'Cod. Materia Modalidade']
num_cols_viz = num_cols.drop(columns=cols_excluir)


media_dos_ultimas_excluidas = num_cols_viz.iloc[:, :-2].mean().mean()


media_general_num_viz = num_cols_viz.mean().mean()



num_cols_viz.hist(bins=10, figsize=(15, 10))
plt.suptitle('Histogramas', y=0.92, fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.grid(axis='y', linestyle='--')
plt.xlim(0, 10)
plt.show()


plt.figure(figsize=(10, 8))
sns.boxplot(data=num_cols_viz)
plt.title('Box Plots')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(0, 11, 1))
plt.ylim(0, 10)
plt.xlabel('MATERIAS') #
plt.ylabel('Mínimo, 2º cuartil, media, 3º cuartil, máximo')


plt.axhline(media_dos_ultimas_excluidas, color='red', linestyle='dashed', linewidth=1, label=f'Media (excluyendo últimas 2 columnas) ({media_dos_ultimas_excluidas:.2f})')


plt.axhline(media_general_num_viz, color='blue', linestyle='dotted', linewidth=1, label=f'Media General ({media_general_num_viz:.2f})')


plt.legend()
plt.tight_layout()
plt.show()
```
